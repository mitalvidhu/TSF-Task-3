```python
sns.set(rc={"figure.figsize":(8, 4)})
sns.barplot(x = 'Profit', y = 'Discount', data = original_df.head(10))
```

## When discount is high, the sales are also high but the profit is negative i.e. the company suffers net loss

```python
sns.set(rc={"figure.figsize":(8, 4)})
sns.barplot(x = 'Sales', y = 'Discount', data = original_df.head(10))
```

```python
original_df['Sub-Category'].unique()
```

```python
sns.set(rc={"figure.figsize":(40, 20)})
sns.lineplot(x = 'Profit', y = 'Discount', data = original_df)
plt.xlabel('Profit (in $)', fontsize = 50)
plt.ylabel('Discount Offered', fontsize = 50)
```

## Profit is positive when the discount offered is less than 20%

```python
sns.set(rc={"figure.figsize":(40, 20)})
sns.lineplot(x = 'Sales', y = 'Discount', data = original_df)
plt.xlabel('Sales (in $)', fontsize = 50)
plt.ylabel('Discount Offered', fontsize = 50)
```

## The company should refrain from offering very high discounts to the customers. Although high discounts will enable a lot more sales, the net profit earned is low and sometimes negative
